```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

key = "b02ca1f46d644ad38e9732380e4ea268"
endpoint = "https://customer-text.cognitiveservices.azure.com/"

def authenticate_client():
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, credential=ta_credential)
    return text_analytics_client

client = authenticate_client()
```

```python
def key_phrase_extraction(client, df):
    train_data = pd.DataFrame(df.reset_index()[['Theme']])
    responses = client.extract_key_phrases(documents = list(df['Text']))
    responses = [r.key_phrases for r in responses]
    all_words = []
    for res in responses:
        for w in res:
            if w not in all_words:
                all_words.append(w)
    keywords_bag = np.array([[1 if w in res else 0 for w in all_words] for res in responses])
    keyphrases = pd.DataFrame(columns=all_words, data=keywords_bag)
    if len(keyphrases) == len(train_data):
        train_data = pd.concat([train_data, keyphrases], axis=1)
        return train_data
    else:
        print('error ' + len(responses) - len(train_data))
    
        
train_data = key_phrase_extraction(client, customer.iloc[:10])
x = train_data.drop(columns=['Theme'])
y = pd.get_dummies(train_data['Theme'])
display(x)
display(y)
```
